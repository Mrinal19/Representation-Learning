```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

audio_fpath = "TeddyPicker.wav"
```

```python
x, sr = librosa.load(audio_fpath, sr=44100)

print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
import librosa
librosa.feature.zero_crossing_rate(Xdb)
```

```python
import numpy as np

chroma_med = librosa.decompose.nn_filter(Xdb, aggregate=np.median, metric='cosine')
chroma_mean = librosa.decompose.nn_filter(Xdb, aggregate=np.average)
```

```python
print(chroma_med)
print(chroma_mean)
```
